```python
from PIL import Image
import os
import numpy as np


def rename_img_from_dic(dic_name, output_path):
    """
    Used for extract single movies from the source dic.
    :param dic_name: Source Dic.
    :param output_path: The output.
    :return:
    """
    path_list = os.listdir(dic_name)
    path_list.sort()
    os.mkdir(dic_name + '/' + output_path)
    for filename in path_list:
        if filename[0] == 'P':
            new_name = filename
            for root, dirs, files in os.walk(dic_name + '/' + filename + '/Red'):
                for file in files:
                    print(os.path.join(root, file))
                    os.renames(str(os.path.join(root, file)), dic_name + '/' + output_path + '/' + new_name + '.tiff')
    return
```

```python
rename_img_from_dic('/Volumes/Muscle/220719/Cas9_CCR5', 'beads_op')
```
